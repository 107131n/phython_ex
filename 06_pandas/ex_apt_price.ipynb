```python
# melt를 이용하여 월별 데이터를 '기간' 칼럼으로 가격은 '평당 분양가격'으로
```

```python
first_df_melt = first_df.melt(id_vars='지역',var_name='기간',value_name='평당 분양가격')
first_df_melt.head(2)
```

```python
# 기간 칼럼의 값을 '연도', '월'로 분리
```

```python
first_df_melt['연도'] = first_df_melt['기간'].str.split('년').str.get(0).astype(int)
```

```python
first_df_melt['월'] = first_df_melt['기간'].str.split('년').str.get(1).str.replace('월','').astype(int)
```

```python
first_df_melt.head()
```

```python
# last_df 전용면적에서 전체만 가지고 와 last_df_prepare로 저장
```

```python
last_df_prepare = last_df[last_df['전용면적'] == '전체'].copy() # 원본은 놔두고 복사함
```

```python
last_df_prepare.head()
```

```python
last_df_prepare.rename(columns ={'지역명':'지역'},inplace=True) # rename: 컬럼의 이름 변경/ inplace: 원본에 적용 여부
```

```python
last_df_prepare.head(2)
```

```python
# concat으로 데이터 합치기
```

```python
first = first_df_melt[['지역','연도','월','평당 분양가격']]
last = last_df_prepare[['지역','연도','월','평당 분양가격']]
```

```python
df = pd.concat([first,last])
```

```python
df.reset_index(drop=True,inplace=True)
```

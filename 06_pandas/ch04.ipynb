```python
fig = plt.figure()
axes1 = fig.add_subplot(1,1,1)
axes1.hist(tips['total_bill'],bins=8,range=(3,55))
axes1.set_title('histogram of total bill')
axes1.set_xlabel('frequency')
axes1.set_ylabel('total bill')
```

```python
scatter_plot = plt.figure()
axes1 = scatter_plot.add_subplot(1,1,1)
axes1.scatter(tips['total_bill'],tips['tip'],s=tips['tip']*10,c='skyblue')
```

```python
box_plot = plt.figure()
axes1 = box_plot.add_subplot(1,1,1)
axes1.boxplot([tips[tips['sex']=='Female']['tip'],
               tips[tips['sex']=='Male']['tip']],
             labels = ['Female','Male'])
box_plot
```

```python
def record_sex(sex):
    if sex == 'Female':
        return 0
    else:
        return 1
```

```python
tips['sex_color'] = tips['sex'].apply(record_sex)
```

```python
scatter_plot1 = plt.figure()
axes1 = scatter_plot1.add_subplot(1,1,1)
axes1.scatter(tips['total_bill'],
             tips['tip'],
             tips['size']*10,
             tips['sex_color'],
             alpha = 0.5)
scatter_plot1
```

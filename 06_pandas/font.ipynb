```python
import numpy as np
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
data = np.random.randint(-100,100,50).cumsum() # -100부터 100까지 50개/ cumsum(): 누적합계 구함
data
```

```python
plt.plot(range(50),data,'r') # plt.plot(x축, y축)/ r: 빨강
plt.title('시간별 가격 추이')
plt.ylabel('주식 가격')
plt.xlabel('시간(분)') # 한글 깨짐: 폰트 설정하기(3가지 방법)
```

```python
#1. 폰트 파일로 FontProperties() 생성 후 사용 : 파일 path 필요

import matplotlib.font_manager as fm

path = 'D2CodingBold-Ver1.3.2-20180524.ttf'
fontprop = fm.FontProperties(fname = path, size = 15)

plt.plot(range(50),data,'r')
plt.title('시간별 가격 추이', fontproperties=fontprop)
plt.ylabel('주식 가격', fontproperties=fontprop)
plt.xlabel('시간(분)', fontproperties=fontprop)
```

```python
#2. matplotlib.rcParams[]로 설정 : 한 세션에서만 적용, 다른 창에서는 적용 안 됨

import matplotlib

print('설정 파일 위치 : ', matplotlib.matplotlib_fname())
print('기본 설정 폰트 : ', plt.rcParams['font.family'])
print('기본 설정 폰트 사이즈 : ', plt.rcParams['font.size'])
```

```python
plt.rcParams['font.family'] = 'Malgun Gothic' #폰트 변경
plt.rcParams['font.size'] = 13
plt.rcParams['axes.unicode_minus'] = False # -가 깨질 때

plt.plot(range(50),data,'r')
plt.title('시간별 가격 추이')
plt.ylabel('주식 가격')
plt.xlabel('시간(분)') 
```
